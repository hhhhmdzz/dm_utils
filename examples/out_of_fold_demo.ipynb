```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from dm_utils.oof import OOF
import warnings
warnings.filterwarnings("ignore")
```

```python
x, y = load_breast_cancer(return_X_y=True, as_frame=True)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

## XGBClassifier

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

xgb = XGBClassifier()
xgb.fit(xtrain, ytrain)
accuracy_score(xgb.predict(xtest), ytest)
```

## dm_utils

xgb

```python
oof = OOF(task='cls', model='xgb')
oof.fit(xtrain, ytrain, record_time=True)
```

```python
accuracy_score(oof.predict(xtest) > 0.5, ytest)
```

```python
oof = OOF(task='cls', model=['xgb', 'xgb', 'lgb', 'lgb', 'cb'])
oof.fit(xtrain, ytrain, record_time=True)
```

```python
accuracy_score(oof.predict(xtest) > 0.5, ytest)
```
